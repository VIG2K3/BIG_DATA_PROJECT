```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
import seaborn as sns
plt.style.use('ggplot')
import warnings
warnings.filterwarnings('ignore')

df = pd.read_csv('database/student_performance_cleaned.csv') 
```

```python
#Label-Encoding for Nominal Attributes
from sklearn.preprocessing import LabelEncoder

nominal_features = ['gender', 'extracurricular', 'internet', 'learningstyle', 'edutech']

# Create one LabelEncoder per column
label_encoders = {}

for col in nominal_features:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    label_encoders[col] = le  # store encoder to decode later if needed

```

```python
#Ordinal Encoding for Ordinal Attributes
from sklearn.preprocessing import OrdinalEncoder

ordinal_features = ['motivation', 'stresslevel', 'discussions', 'resources', 'finalgrade']

# Define order for each ordinal variable
ordinal_mapping = [
    ['Low', 'Medium', 'High'],            # motivation
    ['Low', 'Medium', 'High'],            # stresslevel
    ['Rarely', 'Often'],                  # discussions
    ['Some', 'Many'],                     # resources
    ['Fail', 'Average', 'Good', 'Excellent']  # finalgrade
]

ordinal_encoder = OrdinalEncoder(categories=ordinal_mapping)
df[ordinal_features] = ordinal_encoder.fit_transform(df[ordinal_features])
```

```python
#Convert all Numerical Columns to integers
df = df.astype(int)
```
