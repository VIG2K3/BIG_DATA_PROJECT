```python
#Identify Categorical and Numerical Data

# Get numeric columns
num_cols = df.select_dtypes(include="number").columns.tolist()

# Remove the index column if it exists
if "index" in num_cols:
    num_cols.remove("index")

# Get categorical columns
cat_cols = df.select_dtypes(include=['object']).columns.tolist()

# Print results
print("Categorical Variables:")
print(cat_cols)
print("Numerical Variables (index removed):")
print(num_cols)

```

```python
# ==== Univariate Analysis ====
# Numeric

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Set Visual Style
sns.set(style="whitegrid", palette="pastel")

# Define Age Bins (18â€“29) with inclusive upper bound on the last bin
age_bins = [18, 20, 23, 26, 29]
age_labels = ["Late Teens (18-20)", "Early 20s (21-23)", "Mid 20s (24-26)", "Late 20s (27-29)"]
df["age_binned"] = pd.cut(df["age"], bins=age_bins, labels=age_labels, include_lowest=True, right=True)

for col in num_cols:
    print(f"\nColumn: {col}")
    print('Skew:', round(df[col].skew(), 2))
    
    plt.figure(figsize=(15,5))

    # ===== Histogram =====
    plt.subplot(1,2,1)
    if col == "age":
        # Keep correct order of bins
        age_counts = df["age_binned"].value_counts().reindex(age_labels)
        age_counts.plot(kind="bar", color='skyblue')
        plt.ylabel("Count")
        plt.title("Histogram of Age (Binned)")
        plt.xticks(rotation=0, ha='center')  # horizontal labels
    else:
        df[col].hist(grid=False, color='skyblue')
        plt.ylabel("Count")
        plt.title(f"Histogram of {col}")

    # ===== Box Plot =====
    plt.subplot(1,2,2)
    if col == "age":
        sns.boxplot(x=df[col], color='pink', 
                    medianprops={'color':'red','ls':':','lw':2})
        plt.title("Box Plot of Age")
        plt.xlim(df[col].min()-1, df[col].max()+1)  # full numeric range with padding
    else:
        sns.boxplot(x=df[col], color='pink', 
                    medianprops={'color':'red','ls':':','lw':2})
        plt.title(f"Box Plot of {col}")

    plt.tight_layout()
    plt.show()
```

```python
# === Univariate Analysis (Categorical) ===

sns.set(style="whitegrid")

for col in cat_cols:
    # Get consistent category order
    cat_order = df[col].value_counts().index
    
    # Create one shared palette mapped to that order
    colors = sns.color_palette("Set2", len(cat_order))
    
    plt.figure(figsize=(12, 5))

    # Bar Plot
    plt.subplot(1, 2, 1)
    sns.countplot(
        x=col, 
        data=df, 
        order=cat_order,         # ensures same order
        palette=colors           # match colors
    )
    plt.title(f"Count Plot of {col}")
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.xticks(rotation=30, ha='right')

    # Pie Chart
    plt.subplot(1, 2, 2)
    df[col].value_counts().loc[cat_order].plot.pie(
        autopct='%1.1f%%',
        startangle=90,
        colors=colors           # exact same palette
    )
    plt.title(f"Pie Chart of {col}")
    plt.ylabel("")

    plt.tight_layout()
    plt.show()
```
