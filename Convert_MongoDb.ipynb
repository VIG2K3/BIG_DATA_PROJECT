```python
# Convert Excel into MongoDb
client = MongoClient("mongodb://localhost:27017/")

db = client["StudentDatabase"]
collection = db["StudentPerformance"]

df = pd.read_excel("C:/Users/VIGNEESH GPL/Desktop/Cleaned_Student_Data.xlsx")
data = df.to_dict(orient="records")
collection.insert_many(data)

print(f"Inserted {len(data)} records into MongoDB.")
```

```python
count = collection.count_documents({})
print(f"Total records in MongoDB: {count}")
```

```python
# Remove Duplicates dataset
seen = set()
duplicates = []

for doc in collection.find():
    doc_copy = dict(doc)
    doc_copy.pop("_id", None)  # remove _id from hash
    key = hashlib.md5(str(doc_copy).encode()).hexdigest()  # hash full document

    if key in seen:
        duplicates.append(doc["_id"])
    else:
        seen.add(key)

if duplicates:
    collection.delete_many({"_id": {"$in": duplicates}})
    print(f"Removed {len(duplicates)} duplicate records.")
else:
    print("No duplicates found.")

count = collection.count_documents({})
print(f"Total records in MongoDB: {count}")
```
